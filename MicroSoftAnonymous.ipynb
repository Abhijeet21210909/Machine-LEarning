# Anonymous-msweb


```python
import pandas as pd
import numpy as np
# Managing Data
d = {'pt1':[],'pt2':[],'pt3':[]}
with open('anonymous-msweb.data', 'r') as f:
    f = f.readlines()
    for i in f:
        i = i.split(',')
        if len(i)==3:
            d['pt1'].append(i[0])
            d['pt2'].append(i[1])
            i[2] = i[2].replace('\n','')
            d['pt3'].append(i[2])
data = pd.DataFrame(d)
# Giving value of x and y
x = data.iloc[:, [1,2]].values
y = data.iloc[:, 0].values
```

```python
# converting string to binary
from sklearn.preprocessing import LabelEncoder
lrx = LabelEncoder()
lry = LabelEncoder()
x[:, 0] = lrx.fit_transform(x[:,0])
x[:, 1] = lrx.fit_transform(x[:,1])
y = lry.fit_transform(y)
```

```python
# Model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
# training testing and splitting
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
model.fit(xtrain,ytrain)
```

```python
# predicting value
ypred = model.predict(xtest)
valueoftest = lry.inverse_transform(ypred)
# trying own value
val = [[1,1]]
vlu = model.predict(val)
vlu = lry.inverse_transform(vlu)
```

```python
# prediction accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ytest,ypred)
accuracy_percent = accuracy*100
print(f"{accuracy_percent}%")
# 99.9969552111561
```
